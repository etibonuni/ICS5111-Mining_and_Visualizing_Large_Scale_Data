```python
#copyr = report.copy()
def redundancyPruning(freqItems):
    items = freqItems.items()
    for item in items:
        if len(item[0])>1:
            support = item[1]
            for keyItem in item[0]:
                singlekey = frozenset([keyItem])
                single = report.get(singlekey)
                if single != None:
                    print item, "->", singlekey, ":", support, "<->", single, "=", single-support 
                    report.update({singlekey:single-support})

    #print len(report)
    items = report.items()
    for item in items:
        if item[1]<=3:
            report.pop(item[0])
            
    return freqItems

print len(report) 
redundancyPruning(report)
print len(report)
```
